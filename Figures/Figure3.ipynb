```python
import numpy as np
import matplotlib.pyplot as plt

# teady states
u1 = 0.5
u2 = 2.0
u3 = 7.5

#spatial variable
z = np.linspace(-20, 20, 500)

#smooth wave profile (tanh shape)
U = (u3 + u1)/2 - (u3 - u1)/2 * np.tanh(z/5)


#plot graph
plt.figure(figsize=(8,5))
plt.plot(z, U, color="black", linewidth=2)

# horizontal lines
plt.axhline(u1, linestyle="--", color="black")
plt.axhline(u2, linestyle="--", color="black")
plt.axhline(u3, linestyle="--", color="black")

# labels
plt.text(15, u1, "u1", va="center")
plt.text(15, u2, "u2", va="center")
plt.text(15, u3, "u3", va="center")

plt.axvline(0, color="black", linewidth=1)

plt.xlabel("z")
plt.ylabel("U")
plt.title("Traveling Wave Profile")

plt.grid(False)

plt.show()
```
