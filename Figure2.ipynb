```python
import numpy as np
import matplotlib.pyplot as plt

#roots for schematic curve
u1 = 1.5
u2 = 3.6
u3 = 6.0

u = np.linspace(0, 10, 1000)

# schematic growth curve with three roots
f = -0.08 * u * (u - u1) * (u - u2) * (u - u3)

plt.figure(figsize=(8, 5))

plt.plot(u, f, color="black", linewidth=2, label="f(u)")
plt.axhline(0, color="black", linewidth=1)

#A1 region
mask1 = (u >= u1) & (u <= u2)
plt.fill_between(u[mask1], f[mask1], 0, alpha=0.45, label="A1")

#A3 region
mask2 = (u >= u2) & (u <= u3)
plt.fill_between(u[mask2], f[mask2], 0, alpha=0.45, label="A3")

#plot graph with legned
plt.text((u1 + u2)/2, -0.55, "A1", fontsize=13, ha="center")
plt.text((u2 + u3)/2, 0.35, "A3", fontsize=13, ha="center")

for point, label in zip([u1, u2, u3], ["u1", "u2", "u3"]):
    plt.plot(point, 0, "ko")
    plt.text(point, -0.13, label, ha="center", fontsize=12)

plt.xlim(0, 6.7)
plt.ylim(-1.5, 2.5)
plt.xlabel("u")
plt.ylabel("f(u)")
plt.title("Area Comparison for Wave Direction")
plt.legend()

plt.show()
```
